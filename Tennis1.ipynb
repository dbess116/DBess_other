```python
####### UPSETS FOR 1968-1979 #######

tennis_1968 = pd.read_csv("../tennis_atp/atp_matches_1968.csv")
tennis_1969 = pd.read_csv("../tennis_atp/atp_matches_1969.csv")
tennis_1970 = pd.read_csv("../tennis_atp/atp_matches_1970.csv")
tennis_1971 = pd.read_csv("../tennis_atp/atp_matches_1971.csv")
tennis_1972 = pd.read_csv("../tennis_atp/atp_matches_1972.csv")
tennis_1973 = pd.read_csv("../tennis_atp/atp_matches_1973.csv")
tennis_1974 = pd.read_csv("../tennis_atp/atp_matches_1974.csv")
tennis_1975 = pd.read_csv("../tennis_atp/atp_matches_1975.csv")
tennis_1976 = pd.read_csv("../tennis_atp/atp_matches_1976.csv")
tennis_1977 = pd.read_csv("../tennis_atp/atp_matches_1977.csv")
tennis_1978 = pd.read_csv("../tennis_atp/atp_matches_1978.csv")
tennis_1979 = pd.read_csv("../tennis_atp/atp_matches_1979.csv")

tennis_1968_1979 = tennis_1970.append([tennis_1969, tennis_1970, tennis_1971, tennis_1972, tennis_1973, tennis_1974, tennis_1975,
                                       tennis_1976, tennis_1977, tennis_1978, tennis_1979])
tennis_1968_1979 = tennis_1968_1979[tennis_1968_1979.tourney_name.str.contains("Davis") == False]
tennis_1968_1979 = tennis_1968_1979[tennis_1968_1979.tourney_name.str.contains("Olympics") == False]
tennis_1968_1979 = tennis_1968_1979.drop(['tourney_id','tourney_name','tourney_level','draw_size','round','match_num','winner_ht'
                                          ,'loser_ht','winner_ioc','loser_ioc','w_svpt','l_svpt','w_SvGms','l_SvGms','w_1stWon'
                                          ,'l_1stWon','w_2ndWon','l_2ndWon','w_df','l_df','minutes','winner_rank_points'
                                          ,'loser_rank_points','winner_entry','loser_entry','w_SvGms','l_SvGms','winner_id'
                                          ,'loser_id','w_1stIn','l_1stIn'], axis = 1)       #omits irrelevant columns#
upsets_1968_1979 = tennis_1968_1979[tennis_1968_1979['winner_rank'] - 300 > tennis_1968_1979['loser_rank']]
#selects only largest upsets of 300+ rank difference#
upsets_1968_1979 = upsets_1968_1979[upsets_1968_1979['loser_rank'] <= 100]      #restricts upsets to losers with top 100 rank#
upsets_1968_1979 = upsets_1968_1979.drop(['tourney_date','winner_age','winner_seed','winner_hand','winner_rank','loser_seed',
                                          'loser_name','loser_hand','loser_age','loser_rank','score','best_of','w_ace',
                                          'w_bpSaved','w_bpFaced','l_ace','l_bpSaved','l_bpFaced'], axis = 1)
                                                                    #omits irrelevant columns#
upsets_1968_1979 = upsets_1968_1979.groupby(['surface']).count()
upsets_1968_1979.columns = ['Total Upsets 1968 - 1979']   #formats columns name#
upsets_1968_1979
```
