```python
# define function for District ID (NCESID)
def id_NCESID(district):
    if district == 'DistrictX':
        return '101010'
    else:
        return '373737'

# define function for TestDate, TestTypeName, and TestSubjectName
def get_TestDate_TestTypeName_TestSubjectName(num_tests):
    if num_tests == 3:
        ddict['TestDate'].append('4/1/2013'); ddict['TestDate'].append('5/1/2013'); ddict['TestDate'].append('6/1/2013')
        ddict['TestTypeName'].append('MCAS ELA'); ddict['TestTypeName'].append('MCAS Math'); ddict['TestTypeName'].append('MCAS Science')
        ddict['TestSubjectName'].append('ELA'); ddict['TestSubjectName'].append('Math'); ddict['TestSubjectName'].append('Science')
    elif num_tests == 2:
        ddict['TestDate'].append('4/1/2013'); ddict['TestDate'].append('5/1/2013')
        ddict['TestTypeName'].append('MCAS ELA'); ddict['TestTypeName'].append('MCAS Math')
        ddict['TestSubjectName'].append('ELA'); ddict['TestSubjectName'].append('Math')
    else:
        ddict['TestDate'].append('6/1/2013')
        ddict['TestTypeName'].append('MCAS Science')
        ddict['TestSubjectName'].append('Science')

# define function for casting the Performance Levels for MCAS scores
def performance_level_casting(col_value):
    if (col_value == 'F'):
        ddict['Score1Value'].append('1-F')
    elif (col_value == 'W'):
        ddict['Score1Value'].append('2-W')
    elif (col_value == 'NI'):                      # define function for mapping performance levels
        ddict['Score1Value'].append('3-NI')
    elif (col_value == 'P'):
        ddict['Score1Value'].append('4-P')
    elif (col_value == 'A'):
        ddict['Score1Value'].append('5-A')
    else:
        ddict['Score1Value'].append('6-P+')
```
